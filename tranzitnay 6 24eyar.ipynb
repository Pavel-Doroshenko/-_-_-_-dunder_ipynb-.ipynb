```python
analiz = pd.concat([oktubre, septiembre, agosto, julio, junio, may, april, march, januyry, february, november, deсember], ignore_index=True)
analiz
```

```python
analiz['Отчетный_период'].replace({202407: '07 июль2024', 202408: '08 август2024', 202409:'09 сентябрь2024', 202401:'01 январь2024', 202402:'02 февраль2024',
                                   202410:'10 октябрь2024', 202406:'06 июнь2024', 202404:'04 апрель2024', 202405:'05 май2024', '01.03.2024':'03 март2024',
                                   202411: '11 ноябрь2024', 202412: '12 декабрь2024'}, inplace=True)
analiz.head(5)
```

```python
analiz = analiz.dropna()
analiz
```

```python
analiz =analiz.rename(columns = {'Способ формирования начислений':'Способ_формирования_начислений'})
analiz
```

```python
analiz['Способ_формирования_начислений'].replace({'ФП':'Фактический объем', 'РО':'Расчетный объем',
                                   'КД':'Коррекция начислений', 'ТП':'Бесприборный учет'}, inplace=True)
analiz.head(5)
```

```python
analiz['Квартира'] = analiz['Квартира'].astype(int)
```

```python
analiz.info()
```

```python
analiz1 = analiz.pivot_table(index=['Квартира', 'Способ_формирования_начислений'], columns=['Отчетный_период'], values='Начислено_кВт*ч',
                             aggfunc="min", fill_value='-')
analiz1
```
